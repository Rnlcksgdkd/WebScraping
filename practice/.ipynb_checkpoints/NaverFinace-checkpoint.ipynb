```python




# 베이스 링크에 따라 전체 일자별 환율을 리스트 형태로 저장
def get_country_exchange(country_link):
    
    date = []
    exchange = []
    page = 0
    
    while(1):
        page += 1
        link = country_link + "&page=" + str(page)

        if get_page(link , date , exchange):
            break
    
    return date , exchange

# 각 Country 환율 html 안에서 날짜/환율을 파싱해서 리턴
def get_page(link , date , exchange):
   
    html = urlopen(link)
    bsObject = BeautifulSoup(html, "html.parser")
    bs = bsObject.select('div.tb1_area tbody tr')[0]

    if len(bs) == 0:  return True
    
    for i,temp in enumerate(bs):
        if i < 2: continue
        date.append(bs[i].select("td")[0].get_text())
        exchange.append(float(bs[i].select("td")[1].get_text().replace(',', '')))
    
    return False
```
