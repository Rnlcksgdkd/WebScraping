```python

## (1). 오크타운 신제품들 이름/가격 추출하기 
# v 1.1 

from urllib.request import urlopen
from bs4 import BeautifulSoup


# 책의 상세 웹페이지 주소를 추출하여 리스트에 저장합니다.
item_urls = []
item_list = []

page = 0
base_link = 'http://orctown-store.com/category/new-products/4/'
end_flag = 1

while(1):
    page += 1
    end_flag = 1
    
    link = base_link + "?page=" + str(page)
    html = urlopen(link)
    bsObject = BeautifulSoup(html, "html.parser")
  
    for cover in bsObject.find_all('li'):
        
        cover1 = cover.select('div' , {'class' : 'thumbnail'})
        if cover1:
            end_flag = 0
            cover2 = cover.find_all('span' , {'style' :"font-size:12px;color:#008BCC;font-weight:bold;" })
            item_list.append([cover1[0].select('img')[0].get('alt') ,cover2[1].string ])
             
    if end_flag: break
        
for item in item_list:
    print(item[0] + " : " + item[1])
    
```

```python

## (2). 오크타운 검색어로 검색하기

item_base_link = "http://orctown-store.com"
search_base_link = "http://orctown-store.com/product/search.html?banner_action=&keyword="
search = input()
item_urls = []
item_list = []
page = 0

while(page < 100):
    page += 1
    link = search_base_link + search + "&page=" + str(page)
    html = urlopen(link)
    bsObject = BeautifulSoup(html, "html.parser")
    
    end = 1
    for cover in bsObject.find_all('div', {'class':'thumbnail'}):
        end = 0
        link = cover.select('a')[0].get('href')
        item_urls.append(item_base_link + link)
    if end == 1: break
#print(item_urls)


for link in item_urls:
    html = urlopen(link)
    bsObject = BeautifulSoup(html, "html.parser")
    item = []
    
    for cover in bsObject.find_all('meta' , {'property':"og:title"}):
        item.append(cover)
    
    for cover in bsObject.find_all('meta' , {'property':'product:sale_price:amount'}):
        item.append(cover)
    item_list.append(item)

print("전체 검색 결과 : " , len(item_list))    
for item in item_list:
    print(item)
```
