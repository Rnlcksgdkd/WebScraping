```python
## (1). 트롤 트레이더스 블랙 프라이데이 물건 서치

item_base_link = "https://thetrolltrader.com"
page = 0
item_lst = []

while(page < 124):
    page += 1
    link = "https://thetrolltrader.com/collections/black-friday-2020" + "?page=" + str(page)
    html = urlopen(link)
    bsObject = BeautifulSoup(html, "html.parser")

    for cover in bsObject.find_all('div' , {"class" : "product-top"}):
        name  = cover.select('a')[0].get('title')
        link = item_base_link + cover.select('a')[0].get('href')
        
        html = urlopen(link)
        bsObject = BeautifulSoup(html, "html.parser")
        price = bsObject.find_all('meta' , {"property" : "og:price:amount"})[0].get('content')
        price = str(price) + " £"
        item_lst.append([name + " (" + price + ")" , link])

for item in item_lst:
    print(item[0] + " -- " + item[1])
```

```python

## (2). 트롤 트레이더스 새 제품 리스트들 보기

# 업데이트 : 2020_12_16

from urllib.request import urlopen
from bs4 import BeautifulSoup

# 최대 페이지 설정
max_page = 5


base_link = "https://thetrolltrader.com"
page = 0
item_list = []

while(page < max_page):
    
    # 새로운 제품 페이지 url
    page += 1
    link = "https://thetrolltrader.com/collections/new" + "?page=" + str(page)
    html = urlopen(link)
    bsObject = BeautifulSoup(html, "html.parser")
    
    cnt = 0
    for cover in bsObject.find_all('a', {'class' : "product-link js-product-link"}):
        cnt += 1
        if cnt%2 == 0:
            #print(cover)
            name = cover.get('title')
            link = cover.get('href')
            span = cover.select('span')[0].select('span')
            if span:
                price = span[0].get_text()
            else:
                price = "Sold Out"
            
            print(name + " (" + price + ")" + " -- " + base_link + link)
            
            
```
