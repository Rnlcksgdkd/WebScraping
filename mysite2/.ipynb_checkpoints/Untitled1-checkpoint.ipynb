```python

## 본격 코드 시작

# INPUT : 숙소 이름 / 숙소 주소
NAME = ["호텔 유리앤" , "역삼 호텔 디 아티스트" , "호텔 인 나인 강남" , "인터컨티넨탈 서울 코엑스" , "호텔 리베라"]
ADDRESS = ["서울 강남구 삼성동 158" , "서울 강남구 역삼동 678-13" , "서울 강남구 삼성동 91-28" ,"서울 강남구 삼성동 159", "서울 강남구 청담동 53-7"]

```

```python
# 숙소 주소 -> 위도/경도 변환
import requests
X = []
Y = []
headers = {
    "Authorization": "KakaoAK 8b4d8c2ac3a501ee0cf3519414386ca2"
}

for i,addr in enumerate(ADDRESS):
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(addr)
    places = requests.get(url, headers = headers).json()['documents']
    
    print("{}(여기어때) ,  {}(KakaoAPI)".format(NAME[i] , places[0]["place_name"] ))
    X.append(places[0]["x"])
    Y.append(places[0]["y"])

import pandas as pd
df = pd.DataFrame({"X" : X , "Y" : Y})
df
```

```python
df["X"] = pd.to_numeric(df["X"])
df["Y"] = pd.to_numeric(df["Y"])

print( df["X"].mean() , df["Y"].mean() ) 
```

```python
import folium
import pandas as pd
df = pd.DataFrame({"X" : X , "Y" : Y})
map_searching = folium.Map(location = [df["Y"].mean() , df["X"].mean()], zoom_start = 13)

for i in range(len(ADDRESS)):
    folium.Marker((Y[i],X[i]) , radius = 10 , color = "red" , popup = NAME[i]).add_to(map_searching)


map_searching
```

```python
save_html = map_searching.save('map.html')
```

```python
import pandas as pd
import folium

# HTML 로 저장한다
def get_Map(NAME, Y, X):
    save_dir = "./"

    df = pd.DataFrame({"X" : X , "Y" : Y})
    df["X"] = pd.to_numeric(df["X"])
    df["Y"] = pd.to_numeric(df["Y"])
    map_searching = folium.Map(location = [df["Y"].mean() , df["X"].mean()], zoom_start = 13)

    for i in range(len(NAME)):
        folium.Marker((Y[i],X[i]) , radius = 10 , color = "red" , popup = NAME[i]).add_to(map_searching)
    map_searching.save('map.html')
```

```python
get_Map(NAME , X , Y)
```
